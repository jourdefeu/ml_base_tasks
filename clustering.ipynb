```python
#sklearn clustering

from sklearn.cluster import KMeans

import pandas as pd
import seaborn as sns

df = pd.read_csv('students.csv')
```

```python
df.info()
```

```python
df_cut=df[['Weight', 'Growth', 'Sex']]
df_cut=df_cut.dropna()
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(df_cut[['Weight', 'Growth']])

df_cut['Label']=kmeans.labels_
```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Label')
```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Sex')
```

```python
sum(df_cut['Sex']=="мужской") # чтобы не создавалась новая таблица (задействуются мощности компа)
```

```python
sum((df_cut['Sex']=="мужской") & (df_cut['Label']==1)) # подсчет несовпадений кластеризации
```

```python
sum((df_cut['Sex']=="женский") & (df_cut['Label']==0))
```

```python
df_cut=df[['Weight', 'Shoe size']]
df_cut=df_cut.dropna() #для изучения не вдаюсь в подробности пропуска данных и просто удаляю пустые
```

```python
kmeans = KMeans(n_clusters=2, random_state=1)
kmeans.fit(df_cut)

df_cut['Label']=kmeans.labels_

sns.scatterplot(data=df_cut, x='Weight', y = 'Shoe size', hue='Label')
```
